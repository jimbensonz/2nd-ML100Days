```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import datasets, linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score, accuracy_score
```

```python
wine_dataset = datasets.load_wine()
wine_dataset['feature_names']
```

```python
X = pd.DataFrame(wine_dataset['data'], columns=wine_dataset['feature_names'])
X
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, wine_dataset.target, test_size=0.1, random_state=4)

regr = linear_model.LinearRegression()
regr.fit(x_train, y_train)
y_pred = regr.predict(x_test)
y_pred
```

```python
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, y_pred))
```

## Linear Regresion: Boston

```python
boston = datasets.load_boston()
boston.keys()
```

```python
X = pd.DataFrame(boston.data, columns=boston.feature_names)
X
```

```python
Y = pd.DataFrame(boston.target, columns=['target_name'])
Y
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.1, random_state=4)

regr = linear_model.LinearRegression()
regr.fit(x_train, y_train)
y_pred = regr.predict(x_test)
y_pred
```

```python
print("Mean squared error: %.2f"
      % mean_squared_error(y_test, y_pred))
```
